```
from fastai import *
from fastai.vision import *
from fastai.vision import *
```

```
path = '/content/drive/MyDrive/classification/data/train'
```

```
data = ImageDataBunch.from_folder(path, train='.',valid_pct=0.1, ds_tfms = get_transforms(),size = 224)
```

```
data.normalize(imagenet_stats)
```

```
data.show_batch(rows = 3, figsize=(7,6))
```

```
data.classes
```

```
learn = create_cnn(data,models.resnet18,metrics = accuracy)
```

```
learn.fit_one_cycle(20) 
```

```
learn.export('/content/drive/MyDrive/classification/class_stage1_resnet18.pkl')
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
interp.plot_top_losses(16,figsize=(12,13))
```
